```python
import numpy as np 
import pandas as  pd
```

```python
import cv2
from keras.models import load_model
```

```python
model_brand = load_model('200X200_car.h5')
```

```python
model_plate = cv2.CascadeClassifier('indian_license_plate.xml')
```

```python
model_car=cv2.CascadeClassifier('cars.xml')
```

```python
cap_mob = cv2.VideoCapture('http://192.168.43.10:8080/video')
cap =cv2.VideoCapture(0)
```

```python
def brand_predict(img):
    img_1=cv2.resize(img , (200,200))
    img_1 =np.reshape(img_1 , (1,200,200,3))
    p = model_brand.predict(img_1)
    p=p.reshape(50,1)
    from os import listdir
    names = listdir('./imgs/')
    max=p[0]
    index = 0
    b=0
    for i in p:
        if i>max :
            max=i
            b = index
        index = index+1
    brand_text = names[b]
    cv2.putText(img,brand_text, (50, 50), cv2.FONT_HERSHEY_COMPLEX, 1, (0,255,0), 2)
    return img
        
```

```python
def license_plate(photo,crop):
    plate=model_plate.detectMultiScale(photo)
    if len(plate) ==0:
        crap = crop.copy()
        cv2.putText(photo ,'may be car has no plate' ,(100,120),cv2.FONT_HERSHEY_COMPLEX_SMALL ,1 ,(0,0,255),2)
        
    else :
        x1=plate[0][0]
        y1=plate[0][1]
        x2=x1 + plate[0][2]
        y2=y1 + plate[0][3]
        crop = crop[y1:y2 , x1:x2 , :]
        photo = cv2.rectangle(photo, (x1,y1),(x2,y2), [0,255,0], 5 )
        
        
    return photo    

        
```

```python
while True:
    ret, photo = cap_mob.read()
    brand_photo = brand_predict(photo)
    car =  license_plate(brand_photo,brand_photo)
    car =cv2.imresize(car ,(200,200))
    cv2.imshow('hi', car)
    if cv2.waitKey(1) == 13:
            break
        
        
cv2.destroyAllWindows()
```
